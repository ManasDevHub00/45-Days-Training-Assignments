6. Create a list: 
arr = [10, 20, 30, 40, 50, 60] 
Write a function that takes the list as input and returns the list in 
reverse order without using the reverse() method.

```python
arr = [10, 20, 30, 40, 50,60]
def reverse_array(arr):
    return arr[::-1]
reversed_arr = reverse_array(arr)
print("Original array:", arr)
print("Reversed array:", reversed_arr)
```

7.Create a tuple: 

data = (5, 10, 15, 20, 25, 30, 35) 
Write a program to:

○ Count how many elements are divisible by 5 

○ Find the sum of all elements

○ Find the average of the tuple

```python
data = (5,10,15,20,25,30,35)
count = 0
for i in data:
    if i % 5 ==0:
        count += 1
print("Count of numbers divisible by 5:", count)

print("Sum of all numbers in the tuple:", sum(data))

print("Average of the numbers in the tuple:", sum(data)/len(data))

```

8.Create a dictionary: 

students = { 
"Aman": 78, 
"Riya": 92, 
"Kriti": 88, 
"Rahul": 95 
} 

Write a program to: 

○ Find the student with the highest marks 

○ Find the student with the lowest marks
 
○ Print only the students who scored more than 85 marks

```python
students = { 
"Aman": 78, 
"Riya": 92, 
"Kriti": 88, 
"Rahul": 95 
} 

max_score = max(students.values())
for name , marks in students.items():
    if marks == max_score:
        print("Top scorer is", name, "with marks", marks)

min_score = min(students.values())
for name , marks in students.items():
    if marks == min_score:
        print("Lowest scorer is", name, "with marks", marks)

for name , marks in students.items():
    if marks > 85:
        print(name, "scored above 85 with marks", marks)
```

9. Write a function: 
count_frequency(arr) 
that takes a list as input and prints the frequency of each element. 
Example: 
arr = [1, 2, 2, 3, 1, 4, 2] 
 Output: 
 
 1 -> 2 times 
2 -> 3 times 
3 -> 1 time 
4 -> 1 time 

```python
arr = [1, 2, 3, 4, 5]
def count_frequency(arr):
    freq ={}

    for num in arr:
        if num in freq:
            freq[num] += 1
        else:
            freq[num] = 1
    
    for key, value in freq.items():
        if value == 1:
            print(f"{key} -> {value} time")
        else:
            print(f"{key} -> {value} times")

arr = [1, 2, 2, 3, 1, 4, 2]
count_frequency(arr)
```
